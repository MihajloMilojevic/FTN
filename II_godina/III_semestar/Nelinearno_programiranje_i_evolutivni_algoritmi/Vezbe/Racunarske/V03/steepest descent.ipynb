```python
import numpy as np
```

```python
def steepest_descent(gradf, x0, gamma, epsilon, N):
    x = np.array(x0, dtype="float").reshape(len(x0), 1)
    for _ in range(N):
        gr = gradf(x)
        x -= gamma*gr
        if np.linalg.norm(gr) < epsilon:
            break
    return x
```

```python
def grad(x):
    x = np.array(x).reshape(len(x), 1)
    return 2 * x
```

```python
print(steepest_descent(grad, [15, 165], 0.15, 1e-6, 100))
```

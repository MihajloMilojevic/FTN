## Fibonačijeva metoda

```python
def fib(f, a, b, tol = 1e-6):
    fs = [0, 1, 1]
    while fs[-1] <= (b-a)/tol:
        fs.append(fs[-1] + fs[-2])
    x_1 = a + (fs[-3]/fs[-1]) * (b-a)
    x_2 = a+b-x_1
    for i in range(len(fs)-1, -1, -1):
        if f(x_1) > f(x_2):
            a = x_1
        else:
            b = x_2
        x_1 = a + (fs[i]/fs[-1]) * (b-a)
        x_2 = a+b-x_1
    return x_1 if f(x_1) < f(x_2) else x_2
```

```python
def f1(x):
    return x**2 - np.sin(2*x)
```

```python
a, b = -3, 3
x_opt = fib(f1, a, b)
print(x_opt, f1(x_opt))
```

```python
x = np.arange(-3, 3, 0.1)
y = f1(x)

plt.plot(x, y)
plt.scatter(x_opt, f1(x_opt))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Primer

x1 = np.arange(-5, 5, 0.1)
x2 = np.arange(-5, 5, 0.1)

x1, x2 = np.meshgrid(x1, x2)

y = np.sin(x1) + np.cos(x2)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x1, x2, y)
plt.show()
```

```python
# Zadatak 3

x = np.arange(-10, 10, 0.1)
y = np.sin(x) + 1/25 * x**2

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, y)
plt.show()
```

```python
# Zadatak 4

def ackley(xs, a, b):
    n = len(xs)
    return -a * np.exp(-b * np.sqrt(1/n * np.sum(xs**2, axis=0))) - np.exp(1/n * np.sum(np.cos(2 * np.pi * xs), axis=0)) + a + np.exp(1)

x1 = np.arange(-5, 5, 0.1)
x2 = np.arange(-5, 5, 0.1)

x1, x2 = np.meshgrid(x1, x2)

y = ackley(np.array([x1, x2]), 20, 0.2)

fig = plt.figure()
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(x1, x2, y)
ay = fig.add_subplot(122)
x = np.arange(-5, 5, 0.1)
y = ackley(np.array([x]), 20, 0.2)
ay.plot(x, y)
plt.show()
```
